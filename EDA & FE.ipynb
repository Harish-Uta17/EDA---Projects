```python
# 🧹 STEP 1 — Clean spaces, tabs, and newlines
df['stops'] = df['stops'].str.replace(r'\s+', ' ', regex=True).str.strip()

# 🧾 STEP 2 — Extract 'stop_count' (like 'non-stop', '1-stop', '2+-stop')
df['stop_count'] = df['stops'].str.extract(r'((?:\d+\+?-stop)|(?:non-stop))', expand=False)

# 🛫 STEP 3 — Extract 'via_city' (text after "Via")
df['via_city'] = df['stops'].str.extract(r'Via\s+(.*)', expand=False)
df['via_city'] = df['via_city'].fillna('Direct').str.strip()
```

```python
import pandas as pd

def convert_stops(x):
    if pd.isna(x):   # handle NaN values
        return None
    x = str(x).lower()  
    if 'non' in x:
        return 0
    elif '1' in x:
        return 1
    elif '2' in x:
        return 2
    elif '3' in x:
        return 3
    else:
        return None

# Apply to your column
df['stop_num'] = df['stop_count'].apply(convert_stops)

print(df[['stop_count', 'via_city', 'stop_num']])
```
